- Binary Search and indexing

```python
def searchInsert(nums, target):
    beg = 0
    end = len(nums)-1
    
    
    while beg<=end:
        midpoint = (end+beg)//2
        print(f"{nums[beg:end+1]}")
        print(f"Midpoint Index: {midpoint}")
        if target == nums[midpoint]:
            return True
        elif target < nums[midpoint]:
            end = midpoint-1
        elif target > nums[midpoint]:
            beg = midpoint +1
        print(f"{nums[beg:end+1]}  beg Index: {beg}  end Index: {end}")
        print(f"\n")
    return beg
```

```python
nums = [1,2,4,8,9,10,19,20,21]
```

```python
searchInsert(nums, 0)
```

```python
for n in [x for x in range(max(nums)+2)]:
    print(f"Target = {n} , Result = {(searchInsert(nums, n))}")
```
